1. Plot the rolling 20-day minimum of the Facebook closing price with the pandas plot() method.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fb = pd.read_csv('fb_stock_prices_2018.csv', index_col='date', parse_dates=True)
quakes = pd.read_csv('earthquakes.csv')
```

```python
fb.info()
quakes.info()
```

```python
fb['fb_rolling'] = fb['close'].rolling(20).min()

fb['fb_rolling'].plot()
plt.title("20 day Rolling Minimum of Facebook Closing Prices")
plt.ylabel("Prices")
plt.show()
```

2. Create a histogram and KDE of the change from open to close in the price of Facebook stock.

```python
fb.head()
```

```python
fb['Close_Minus_Open'] = fb['close'] - fb['open']
fig, axes = plt.subplots(figsize=(8,5))
data = fb['Close_Minus_Open']
data.plot(kind='hist', ax=axes, alpha=0.5,
    bins = 35, legend=True,density=True,
    label='Histogram'
    )
data.plot(kind='kde',
          ax=axes,
          label='KDE',
          legend=True
)
plt.title('Histogram and KDE of Daily Change (Close - Open)')
plt.xlabel('Price Change')
```



3. Using the earthquake data, create box plots for the magnitudes of each magType used in Indonesia.





```python
quakes_filtered = quakes.query('parsed_place == "Indonesia"')

quakes_filtered.boxplot(column='mag', by='magType', figsize=(8,5))
plt.title('Box Plot of Magnitudes by Type in Indonesia')
plt.xlabel('Magnitude Type')
plt.ylabel('Magnitude')
plt.show()
```

4. Make a line plot of the difference between the weekly maximum high price and the weekly minimum low price for Facebook. This should be a single
line.



```python
fb.index = pd.to_datetime(fb.index)
weekly_difference = fb.resample('W').agg({'high':'max', 'low':'min'})

weekly_difference['Difference'] = weekly_difference['high'] - weekly_difference['low']
weekly_difference.plot()
plt.title('Weekly Differences of the Max and Min Price')
plt.xlabel('Price Difference')
plt.show()
```

5. Using matplotlib and pandas, create two subplots side-by-side showing the effect that after-hours trading has had on Facebook's stock price:

- The first subplot will contain a line plot of the daily difference between that day's opening price and the prior day's closing price (be sure to review the Time series section of Aggregating Pandas DataFrames for an easy way to do this).

- The second subplot will be a bar plot showing the net effect this had monthly, using resample().

- Bonus #1: Color the bars according to whether they are gains in the stock price (green) or drops in the stock price (red).

- Bonus #2: Modify the x-axis of the bar plot to show the threeletter abbreviation for the month.

```python
fig, axes = plt.subplots(1, 2, figsize=(20,8))

fb['today_open_prior_close'] = fb['close'].shift(1)
fb['after_hours_diff'] = fb['open'] - fb['today_open_prior_close']

fb['after_hours_diff'].plot(ax=axes[0])
axes[0].set_title("Daily After Hours Difference")
axes[0].set_ylabel("Price Difference")


monthly = fb['after_hours_diff'].resample('ME').sum()
colors = ['skyblue' if x > 0 else 'maroon' for x in monthly]
monthly.plot(kind='bar',
             ax=axes[1],
             color=colors)
axes[1].set_title('Net Effect of After-Hours Monthly')
axes[1].set_ylabel('Price Change')
axes[1].set_xticklabels(monthly.index.strftime('%b'), rotation=0)
plt.show()
```
