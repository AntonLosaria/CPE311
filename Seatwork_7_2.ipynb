1. Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows both the number of meteorites and the
95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). Hint:
Be sure to convert the year column to a number as we did in the previous exercise.
2. Using the meteorite data from the Meteorite_Landings.csv file, compare summary statistics of the mass column for the meteorites that
were found versus observed falling.

```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv('/content/drive/MyDrive/CPE311/Sw7.1/Meteorite_Landings.csv')
```

```python
data['year'] = data['year'].str.slice(6,10)
data['year'] = pd.to_numeric(data['year'], errors='coerce')
data['mass (g)'] = pd.to_numeric(data['mass (g)'], errors='coerce')
filtered_data = data[(data['year']>= 2005)&(data['year']<= 2009)]
filtered_data = filtered_data[filtered_data['fall'].isin(['Fell', 'Found'])]
pivotTable = pd.pivot_table(
    filtered_data,
    values='mass (g)',
    index='year',
    columns='fall',
    aggfunc=[
        'count',
        lambda x: x.quantile(0.95)  # 95th percentile
    ]
)
print(pivotTable)
```

```python
summary = data.groupby('fall')["mass (g)"].describe()
print(summary)
```

Exercise Part 5:
Using the taxi trip data in the 2019_Yellow_Taxi_Trip_Data.csv file, resample the data to an hourly frequency based on the dropoff time.
Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount, then find the 5 hours with the most tips.

```python
import numpy as np
import pandas as pd
```

```python
data2 = pd.read_csv('/content/drive/MyDrive/CPE311/SW6.2/2019_Yellow_Taxi_Trip_Data.csv')
```

```python
data2['tpep_dropoff_datetime'] = pd.to_datetime(data2['tpep_dropoff_datetime'])
data2['hour'] = data2['tpep_dropoff_datetime'].dt.floor('h')
Hf = data2.groupby('hour')[['trip_distance','fare_amount','tolls_amount','tip_amount']].sum()
highest = Hf.sort_values('tip_amount', ascending=False).head(5)
print(highest)
```
