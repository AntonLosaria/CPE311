2. Create a program codes that demonstrate how to use dynamic programming to solve the same problem

```python
def fibodp(n):
    if n <= 1:
        return n
    if n == 1 or n == 2:
      return 1
    fibo = [0] * (n+1)
    fibo[0] = 0
    fibo[1] = 1

    for i in range(2, n+1):
        fibo[i] = fibo[i - 1] + fibo[i - 2]

    return fibo[n]
n = 10
print(f"The {n} number is: {fibodp(10)}")
```

In the given dynamic programming, it basically sets up the base and then uses that to solve sub-problems that will build up to solve the main problem.

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]

  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

**Code Analysis**

From what I saw, all of them were able to solve the problem as they all had the same answer. The difference is how they were able to solve all of them. For recursion, it will solve multiple sub-problems even if these subproblems are the same, just in a different order. For bottom-up approach, it does basically does the same but it starts from the base. It starts with the first value and then create combinations from that one, but it doesn't do the same combination twice, as well as storing the past values that will help with comparing with new values that are being added. For top-down approach, it's basically the same as recursion except that it doesn't calculate the same sub-problem or combination twice, it just returns the value which eliminates redudant calculations. The main idea here is that all of these sample programs may solve the problem but if the input data is expanded, picking the right one may help reduce the time needed to solve it.

```python
def rec_budget(w, wt, val, n, memo=None):
  if memo is None:
    memo = {}

  if n == 0 or w == 0:
    return (0, 0)

  key = (n, w)
  if key in memo:
    return memo[key]


  if wt[n-1] > w:

    result = rec_budget(w, wt, val, n-1, memo)
    memo[key] = result
    return result
  else:

    takebene, takeexpe = rec_budget(w - wt[n-1], wt, val, n-1, memo)
    total_benefit_take = val[n-1] + takebene
    total_expense_take = wt[n-1] + takeexpe

    skipbene, skipexpe = rec_budget(w, wt, val, n-1, memo)

    if total_benefit_take >= skipbene:
      result = (total_benefit_take, total_expense_take)
    else:
      result = (skipbene, skipexpe)

    memo[key] = result
    return result

max_benefit, max_expense = rec_budget(budget, expenses, benefit, n)
print(f"Maximum benefit: {max_benefit}")
print(f"Maximum expense: {max_expense}")
```
