Suppose we had a linear regression model with one feature that learned the following bias and weights:

b = 1

w1 = 3

Further, suppose the following feature values for a given example:

x1 = 0

Compute the linear regression prediction for this particular example using python in Google Colab. Make sure to plot the solution.Provide an analysis for the output.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from os import wait
b = 1
w1 = 3
x1 = 0

y = w1*x1 + b

xval = np.linspace(0,9,10)
yval = w1*xval + b

print(f'Prediction of x1=0 -> y = {y}')

plt.plot(xval, yval, label='y = 3Xn + 1')
plt.scatter(x1, y, label='(x1,y)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Analysis

For the linear regression, it has a positive slope as when different inputs are used to form a line in the graph, it steadily increases. The weight also determines how steep the slope is as every increase in the input it is multiplied by the weight.
