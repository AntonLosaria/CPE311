1. Plot the rolling 20-day minimum of the Facebook closing price with the pandas plot() method.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fb = pd.read_csv('fb_stock_prices_2018.csv', index_col='date', parse_dates=True)
quakes = pd.read_csv('earthquakes.csv')
```

```python
fb.info()
quakes.info()
```

```python
fb['fb_rolling'] = fb['close'].rolling(20).min()

fb['fb_rolling'].plot()
plt.title("20 day Rolling Minimum of Facebook Closing Prices")
plt.ylabel("Prices")
plt.show()
```
