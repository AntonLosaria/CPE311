```python
# Median
sorted_salary = salaries.sort()
index = len(salaries)//2
median = salaries[index]
print(median)

#Sort the dataset first and then get the middle index by dividing the length of the dataset and dividing by 2, then use that index to access the median
```

```python
# Mode
salary_count = {}
for salary in salaries:
  salary_count[salary] = salary_count.get(salary, 0) + 1
highest = 0
for count in salary_count.values():
  if count > highest:
    highest = count
if highest == 1:
  print("All items appear once")
mode = []
for salary,count in salary_count.items():
  if count == highest:
    mode.append(salary)
print(f"Mode: {mode} - Frequency: {highest}")

#Through the first loop, getting the count of each salaries and store them in a dictionary. Declare a variable to store the highest count, then use a loop to iterate over the count values of salary_count
#until it stores the highest count. Lastly, display all salary appear once if the mode is 1 and if not, then use a loop to store all salaries that has the same number of count then print the mode.
```

```python
#Variance
average = sum(salaries)/len(salaries)
len_minus_one = len(salaries) - 1
x_xmean_squared = []

for salary in salaries:
  temp = (salary - average)**2
  x_xmean_squared.append(temp)

summation = sum(x_xmean_squared)
variance = summation/len_minus_one
print(variance)

#First getting the average again. Then getting the N - 1 by using len then minus one to that. Declare a variable where it will store all the (x-average)**2.
#Use a loop to calculate each salary of (x-average)**2, then store in the variable declared earlier. Then use the sum function to get the sum of the (x-average)**2.
#Lastly, divide the summation to the len - 1.
```

```python
# Standard Deviation
SD = variance ** 0.5
print(SD)

#Since we already have the variance, we only raise it 0.5 which acts as a square root.
```

```python
# Interquartile range
import numpy as np
Q1 = np.quantile(salaries, 0.25)
Q3 = np.quantile(salaries, 0.75)
IQR = Q3 - Q1
print(IQR)

#Getting the Q1 and Q3 through numpy and simlpy subtracting Q1 from Q3
```

```python
# Quartile coefficient of dispersion
QCD = (Q3 - Q1) / (Q3 + Q1)
print(QCD)

#Since we already have Q3 and Q1 from earlier, simply use the formula for Quartile coefficient of dispersion
```
